```python
import pandas as pd
```

```python
df = pd.read_excel("Sample_Longlist_Data.xlsx")
df.head(10)
```

```python
df.info()
```

```python
df['product_count'] = df['sellerproductcount'].str.extract(r'over\s*([\d,]+)\s*results|([\d,]+)\s*results')[0].fillna(df['sellerproductcount'].str.extract(r'([\d,]+)\s*results')[0]).replace(',', '', regex=True).astype(float)

print("Before transformation:", df['sellerproductcount'].notna().sum())
print("After transformation:", df['product_count'].notna().sum())
```

```python
df['positive_ratings'] = df['sellerratings'].str.extract(r'(\d+) (?:total\s*)?ratings').astype(float)
df['positive_ratings_percentage'] = df['sellerratings'].str.extract(r'(\d+)%').astype(float)

print(df['sellerratings'].notna().sum())
print(df['positive_ratings'].notna().sum())
print(df['positive_ratings_percentage'].notna().sum())
```

```python
df['email'] = df['sellerdetails'].str.extract(r'([\w\.-]+@[\w\.-]+)')
df['phone'] = df['sellerdetails'].str.extract(r'(\+?\d[\d\s()-]{8,}\d)')

print(df['sellerdetails'].notna().sum())
print(df['email'].notna().sum())
print(df['phone'].notna().sum())
```

```python
df['country'] = df['businessaddress'].str.extract(r'(US|DE|CN)')

print(df['businessaddress'].notna().sum())
print(df['country'].notna().sum())
```

```python
def assign_weights(df):
  weights = {
    "positive_ratings": 0.30,
    "product_count": 0.15,
    "positive_ratings_percentage": 0.25,
    "hero_product1_ratings": 0.15,
    "hero_product2_ratings": 0.15,
    "negative_ratings_12m": -0.20,
    "negative_ratings_90d": -0.10,
    "negative_ratings_30d": -0.05
  }

  # Filling null values with zeros
  df['positive_ratings'] = df['positive_ratings'].fillna(0)
  df['product_count'] = df['product_count'].fillna(0)
  df['positive_ratings_percentage'] = df['positive_ratings_percentage'].fillna(0)
  df['Hero Product 1 #ratings'] = df['Hero Product 1 #ratings'].fillna(0)
  df['Hero Product 2 #ratings'] = df['Hero Product 2 #ratings'].fillna(0)
  df['Max % of negative seller ratings - last 30 days'] = df['Max % of negative seller ratings - last 30 days'].fillna(0)
  df['Max % of negative seller ratings - last 90 days'] = df['Max % of negative seller ratings - last 90 days'].fillna(0)
  df['Max % of negative seller ratings - last 12 months'] = df['Max % of negative seller ratings - last 12 months'].fillna(0)

  # Assigning weights to their respective columns
  df['weighted_score'] = (
    df['positive_ratings'] * weights['positive_ratings'] +
    df['product_count'] * weights['product_count'] +
    df['positive_ratings_percentage'] * weights['positive_ratings_percentage'] +
    df['Hero Product 1 #ratings'] * weights['hero_product1_ratings'] +
    df['Hero Product 2 #ratings'] * weights['hero_product2_ratings'] +
    df['Max % of negative seller ratings - last 12 months'] * weights['negative_ratings_12m'] +
    df['Max % of negative seller ratings - last 90 days'] * weights['negative_ratings_90d'] +
    df['Max % of negative seller ratings - last 30 days'] * weights['negative_ratings_30d']
  )

  return df
```

```python
df = assign_weights(df)
df[['weighted_score']].describe()
```

```python
def promising_sellers(df):
  df = df[['sellerlink', 'product_count', 'positive_ratings', 'positive_ratings_percentage', 'email', 'phone', 'country', 'weighted_score']]
  filtered_df = df[df['country'] != 'CN']
  filtered_df = filtered_df[filtered_df['weighted_score'] > 1075.3]  # 1075.3 because we will get top 25% sellers
  return filtered_df
```

```python
filtered_df = promising_sellers(df)
```
